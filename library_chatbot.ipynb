```python
intents = {
    "greeting": {
        "patterns": ["hi", "hello", "hey"],
        "responses": ["Hello! welcome", "Hi there! welcome , how can i help you", "Hey! How can I help you in the library?"]
    },
    "book_query": {
        "patterns": ["book", "available", "library"],
        "responses": ["Please tell me the book name."]
    },
    "book_price" : {
        "patterns" : ["price", "cost", "rate"],
        "responses" : ["The cost of the books ranges from 100 to 500"]
    },
    "book_search": {
      "patterns": ["find book", "search book", "do you have"],
      "responses": ["Please tell me the book name or subject."]
    },
    "library_timing": {
      "patterns": ["timing", "opening time", "closing time"],
      "responses": ["Library is open from 10 AM to 4 PM."]
    },

    "fine_query": {
        "patterns": ["fine", "late", "penalty"],
        "responses": ["Please enter number of late days."]
    },
    "fine_calculation" : {
        "patterns" : ["calculate", "fine", "penalty"],
        "responses" : ["Late fine 50 per day after the due date.", "Penalty = extra days × 50."]
    },
    "goodbye": {
        "patterns": ["bye", "exit"],
        "responses": ["Goodbye!", "See you later!"]
    },
    "thankyou": {
        "patterns": ["thanks", "thankyou"],
        "responses": ["You're welcome!"]
    }
}

# Library books database
books = {
    "python": 7,
    "java": 6,
    "data science": 2,
    "machine learning": 3,
    "dbms": 1
}
book_prices = {
    "python": 300,
    "java": 250,
    "data science": 400,
    "machine learning": 500,
    "dbms": 200
}
```

```python
import random

# Function to detect intent
def get_intent(user_input):
    user_input = user_input.lower()
    for intent, data in intents.items():
        for pattern in data["patterns"]:
            if pattern in user_input:
                return intent
    return "unknown"


# Function to extract book name
def extract_book(user_input):
    user_input = user_input.lower()
    for book in books:
        if book in user_input:
            return book
    return None


def chatbot():
    print("Library Bot: Hello! Welcome to the College Library.")

    while True:
        user_input = input("\nYou: ").lower()
        intent = get_intent(user_input)

        if intent == "greeting":
            print("Bot:", random.choice(intents[intent]["responses"]))

        elif intent == "thankyou":
            print("Bot:", random.choice(intents[intent]["responses"]))

        elif intent == "book_query" or intent == "book_search":
            book_name = extract_book(user_input)

            if book_name:
                print(f"Bot: {book_name} book is available. copies are: {books[book_name]}")
            else:
                print("Bot: Sorry, book not found. Please try another name.")

        # Book price
        elif intent == "book_price":
          book_name = extract_book(user_input)
          if book_name:
            print(f"Bot: Price of {book_name} book is ₹{book_prices[book_name]}")
          else:
            print("Bot: Book not found. Please mention the book name.")

        elif intent == "library_timing":
            print("Bot:", random.choice(intents[intent]["responses"]))

        # Fine calculation
        elif intent == "fine_query" or intent == "fine_calculation":
            try:
                days = int(input("Bot: Enter number of late days: "))
                fine = days * 50   # mathematical calculation
                print(f"Bot: Your total fine is ₹{fine}")
            except:
                print("Bot: Invalid input! Please enter a valid number.")

        # Exit
        elif intent == "goodbye":
            print("Bot:", random.choice(intents[intent]["responses"]))
            break
        else:
            print("Bot: Sorry, I didn't understand that.")
chatbot()
```
